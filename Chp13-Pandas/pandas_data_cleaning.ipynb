```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  # we will use this for plotting

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
median_salary_year = df['salary_year_avg'].median()
```

```python
median_salary_hour = df['salary_hour_avg'].median()
```

```python
df_filled = df

df['salary_year_avg'] = df_filled['salary_year_avg'].fillna(median_salary_year)
df['salary_hour_avg'] = df_filled['salary_hour_avg'].fillna(median_salary_hour)
```

```python
df.loc[20:30]
```

```python
df_unique = df_filled

df_unique = df_unique.drop_duplicates()

print("original length: ", len(df_filled))
print("unique length: ", len(df_unique))
print("number of duplicates: ", len(df_filled) - len(df_unique))
```

```python
df_unique = df_unique.drop_duplicates(subset=['job_title', 'company_name'])

print("original length: ", len(df_filled))
print("unique length: ", len(df_unique))
print("number of duplicates: ", len(df_filled) - len(df_unique))
```
