```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
df.info()
```

```python
df['job_posted_date'] = pd.to_datetime(df.job_posted_date) # Converting to datetime
df.info()
```

```python
df['job_posted_month'] = df.job_posted_date.dt.month # Extracting month from datetime
df['job_posted_year'] = df.job_posted_date.dt.year # Extracting year from datetime
df.info()
```

```python
df
```

```python
df.drop(labels=['job_posted_year'], axis=1, inplace=True) # Dropping the year datetime column
```

```python
df.info()
```

```python
df.dropna(subset=['salary_year_avg'], inplace=True) # Dropping rows with missing job salary
```
