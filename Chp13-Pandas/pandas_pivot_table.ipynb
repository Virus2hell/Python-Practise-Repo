```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  # we will use this for plotting

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.pivot_table(index='job_title_short', aggfunc='size')
```

```python
df.groupby('job_title_short').size()
```

```python
df.pivot_table(index='job_title_short', values='salary_year_avg', aggfunc='median')
```

```python
df.groupby('job_title_short')['salary_year_avg'].median()
```

```python
df.pivot_table(index='job_title_short', values='salary_year_avg', columns='job_country', aggfunc='median')
```

```python
df.pivot_table(index='job_country', values='salary_year_avg', columns='job_title_short', aggfunc='median')
```

```python
top_countries = df['job_country'].value_counts().head(5).index

job_country_salary = df.pivot_table(index='job_country', values='salary_year_avg', columns='job_title_short', aggfunc='median')

job_country_salary = job_country_salary.loc[top_countries]

job_country_salary.plot(kind='bar', figsize=(12, 6))
plt.title('Median Salary by Job Title and Country')
```
