```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  # we will use this for plotting

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
job_counts = df['job_title_short'].value_counts().sort_values()
plt.barh(job_counts.index, job_counts.values)
plt.xlabel('Number of Jobs')
plt.ylabel('Job Title')
plt.title('Job Title Distribution')
```

```python
plt.pie(job_counts.values, labels=job_counts.index, autopct='%1.1f%%')
```
